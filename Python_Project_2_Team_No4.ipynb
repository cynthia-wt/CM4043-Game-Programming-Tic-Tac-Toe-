```python
import random
import numpy as np
from time import sleep

class YNError(Exception):
    def __init__(self, message):
        self.message = message
    def __str__(self):
        return str(self.message)

class InputError(Exception):
    def __init__(self, message):
        self.message = message
    def __str__(self):
        return str(self.message)

class MoveError(Exception):
    def __init__(self, message):
        self.message = message
    def __str__(self):
        return str(self.message)

class ConsecutiveError(Exception):
    def __init__(self, message, row):
        self.message = message
        self.row = row
    def __str__(self):
        return str(self.message + self.row)
```

```python
bold = "\033[1m"
red = '\033[91m'
cyan = '\033[96m'
purple = '\033[95m'
end = "\033[0;0m"

class Game:
    mat = None # this represents the game board
    rows = 0 # this represents the number of rows of the board
    cols = 0 # this represents the number of columns of the board
    turn = 0 # this represents whose turn it is (1 for player 1, 2 for player 2)
    wins = 0 # this represents the number of consecutive disks you need to force in order to win

    def __init__(game):

        # initialize dimension
        if dimension == 1:
            game.rows = game.cols = 5

        else:
            if dimension_no == 1:
                game.rows = game.cols = 3
            if dimension_no == 2:
                game.rows = game.cols = 4

        # initialize playing mode
        global player_2
        if playing_mode == 1:
            player_2 = 'human'

        else:
            player_2 = 'computer'

        # initialize game.wins
        game.wins = consecutive_number
        if game.wins not in range(3,game.rows+1):
            raise ConsecutiveError(bold + red + "\nPlease choose a number from 3 to ", str(game.rows) + end)

        # initialize game board
        game.mat = np.zeros((game.rows,game.cols), dtype=int)


        sleep(1.5)
        # return Game Starts message
        print(bold + '\n\nGame Starts!\n' + end)

        # welcome message
        print("\033[1;37;40m  Welcome to Tic Tac Toe  \n" + end)


    # display the game board
    def display_board(game):
        return(game.mat)
        sleep(0.3)

    ##### for human player

    # check validity of player's move
    def check_move(game, x, y, pop):

    ## for normal move
        if pop_out == 'N':
            if game.mat[x,y] == 0:
                return True
            else:
                print(bold+red + "\nThis cell is occupied" + end)
                return False

    ## for pop out action
        else:

            if x != game.rows-1:
                print(bold+red + "\nOnly can pop bottom row" + end)
                return False

            elif game.mat[x,y] == 0:
                print(bold+red + "\nCell chosen is empty" + end)
                return False

            elif game.mat[x,y] != pop and game.mat[x,y]!= 0:
                print(bold+red + "\nThis move is not acceptable"  + end)
                return False

            else:
                r = 0
                nothing_above = True

                while r in range(game.rows-1):
                    if game.mat[r,y] != 0:
                        nothing_above = False
                        break
                    else:
                        r += 1
                        continue

                if nothing_above == True:
                    print(bold+red + "\nNothing above for pop to occur" + end)
                    return False

            return True

    # apply player's move
    def apply_move(game, x, y, pop):

    ## for normal move
        if pop_out == 'N':
            game.mat[x,y] = pop
            return(bold + '\nMove Completed\n' + end)

    ## for pop out action
        else:
            for row in range(x,-1,-1):
                game.mat[row,y] = game.mat[row-1,y]
            game.mat[0][y] = 0
            return(bold + '\nMove Completed\n' + end)

    #### for computer player

    def computer_move(game):
        
        pop = 2
        ## for normal move
        # Check for empty places on board

        empty = []
        for i in range(len(game.mat)): # run through the rows
            for j in range(len(game.mat)): # run through the columns in the row
                if game.mat[i,j] == 0: # if position == 0, no player has chosen the location yet
                    empty.append((i, j)) # then the location is a possible choice and is added as a tuple in the list

        # random allocation
        current_loc = random.choice(empty) #choose one of the possibilities randomly
        game.mat[current_loc] = pop

        empty = []
        return(bold + 'Computer\'s Turn\n' + end)

    # check for victory
    def check_victory(game):

        # set winning combinations
        horizontal = []
        vertical = []
        diagonal = []

        def count_consec(list):
            count = 0
            win = False
            for n in list:
                if n == 1:
                    count += 1
                    if count == game.wins:
                        win = True
                        break
                    else:
                        continue
                else:
                    count = 0
                    
            if win == True:
                return True
            else:
                return False

        # check for horizontal win
        
        for pop in [1,2]:
            for m in range(game.rows):
                for n in range(game.cols):
                    if game.mat[m,n] == pop:
                        horizontal.append(1)
                    else:
                        horizontal.append(0)

                if count_consec(horizontal) == True:
                    if pop == 1:
                        return(1)
                    else:
                        return(2)
                else:
                    horizontal = []
                
        # check for vertical win
        for pop in [1,2]:
            for n in range(game.cols):
                for m in range(game.rows):
                    if game.mat[m,n] == pop:
                        vertical.append(1)
                    else:
                        vertical.append(0)  
                if count_consec(vertical) == True:
                    if pop == 1:
                        return(1)
                    else:
                        return(2)
                else:
                    vertical = []
                
        # check for diagonal win

        if dimension_no == 1:
            diagonal_list1 = [(0,0)]
            diagonal_list2 = [(0, game.cols-1)]

        if dimension_no == 2:
            diagonal_list1 = [(0,0),(0,game.cols-3),(game.rows-3,0)] # +=1, +=1
            diagonal_list2 = [(0, game.cols-1),(0,game.cols-2),(game.rows - 3,game.cols - 1)] # +=1, -=1

        if dimension == 1:
            diagonal_list1 = [(0,0),(0,game.cols-3),(game.rows-3,0),(0,game.cols - 4),(game.rows - 4,0)]
            diagonal_list2 = [(0, game.cols-1),(0,game.cols-2),(game.rows - 3,game.cols - 1),(0,game.cols-3),(game.rows - 4,game.cols - 1)]

        
        for pop in [1,2]:
            for i in diagonal_list1:
                a = i[0]
                b = i[1]
                diagonal = []
                while a in range(game.rows) and b in range(0, game.cols):
                    if game.mat[a,b] == pop:
                        diagonal.append(1)
                        a += 1
                        b += 1
                        continue
                    else:
                        diagonal.append(0)
                        a += 1
                        b += 1
                        continue
                if count_consec(diagonal) == True:
                    if pop == 1:
                        return(1)
                    else:
                        return(2)
                else:
                    diagonal=[]
                    
            for i in diagonal_list2:
                a = i[0]
                b = i[1]
                diagonal = []
                while a in range(game.rows) and b in range(0, game.cols):
                    if game.mat[a,b] == pop:
                        diagonal.append(1)
                        a += 1
                        b -= 1
                        continue
                    else:
                        diagonal.append(0)
                        a += 1
                        b -= 1
                        continue
                if count_consec(diagonal) == True:
                    if pop == 1:
                        return(1)
                    else:
                        return(2)
                else: 
                    diagonal=[]

        # check for tie
        used_cells = 0

        for row in game.mat:
            for cell in row:
                if cell != 0:
                    used_cells += 1

        if used_cells == game.rows * game.cols:
            return(0)

    def reset(game):
        game.mat =  np.zeros((game.rows,game.cols), dtype=int)
```

```python
def menu():

    print("\n\033[1;37;40m  Please Choose Your Game Setting  \n" + end)
    # choose the playing mode
    global playing_mode
    playing_mode = int(input(bold + cyan +"Mode of Playing: (1) Multiplayer Vs (2) Computer "+ end + bold + "\nYour Choice: " + end))
    sleep(0.3)
    if playing_mode != 1 and playing_mode != 2:
        raise InputError(bold+red + "\nPlease enter 1 or 2" + end)
        sleep(0.3)

    # enter player's name
    global player1
    player1 = input(bold + "\nPlease enter Player 1's name: " + end)

    if playing_mode == 1:
        global player2
        player2 = input(bold + "Please enter Player 2's name: " + end)

    # choose the dimension for the board
    global dimension
    dimension = int(input(bold + cyan + "\nDimension: (1) Default (2) Custom " + end + bold + "\nYour Choice: " + end))
    sleep(0.3)
    if dimension != 1 and dimension != 2:
        raise InputError(bold+red + "\nPlease enter 1 or 2" + end)
        sleep(0.3)

    global dimension_no
    dimension_no = 5
    if dimension == 2:
        dimension_no = int(input(bold + cyan + "\nChoose a dimension: (1) 3x3 (2) 4x4 " + end + bold + "\nYour Choice: " + end))
        sleep(0.3)
        if dimension_no != 1 and dimension_no != 2:
            raise InputError(bold+red + "\nPlease enter 1 or 2" + end)
            sleep(0.3)

    # choose the number of consecutive 1 or 2 required for a win
    global consecutive_number
    if dimension_no == 1:
        consecutive_number = 3
    else:
        consecutive_number = int(input(bold+cyan+" \nNumber of consecutive 1 or 2 required for a win:" + end + bold + "\nYour Choice: " + end))
        sleep(0.3)

    return cyan + bold +" \n.....Starting Up Game....." + end

def input_menu(pop):

    # choose whether if you want to pop out
    if pop == 1:
        player = player1
    else:
        player = player2

    print(bold + "\n" + player + '\'s Turn' + end)

    global pop_out
    pop_out= (input(bold+cyan + "Pop Out (Y/N): " + end)).upper()
    if pop_out != 'Y' and pop_out != 'N':
        raise YNError(bold+red + "\nPlease enter Y or N" + end)
        print()

    # choose x and y values
    if pop == 1:
        global player_1_x
        player_1_x = int(input(bold + "\nRow Number: " + end))
        if player_1_x not in range(my_game.cols):
            raise IndexError
        global player_1_y
        player_1_y = int(input(bold + "Column Number: " + end))
        if player_1_y not in range(my_game.rows):
            raise IndexError

    else:
        global player_2_x
        player_2_x = int(input(bold + "\nRow Number: " + end))
        if player_2_x not in range(my_game.cols):
            raise IndexError
        global player_2_y
        player_2_y = int(input(bold + "Column Number: " + end))
        if player_2_x not in range(my_game.rows):
            raise IndexError

def play_again():

    while True:
        global continueplay
        continueplay = (input(bold + "\nWould you like to play again? (Y/N)\n " + end)).upper()

        if continueplay == "Y":
            my_game.turn = 1
            print()
            sleep(0.3)
            my_game.reset()
            print(bold + '\n\033[1;37;40mWelcome to Tic Tac Toe!\n' + end)
            return True
        elif continueplay == "N":
            sleep(0.3)
            return True
        else:
            print(bold + red + "\nPlease enter Y/N \n"  + end)
            continue
```
